```python
import pandas as pd
import numpy as np
from collections import Counter
import requests
import os
import json
```

# Yelp Data

```python
yelp_data=[]
with open('data/yelp_academic_dataset_business.json','r') as f:
    for line in f:
        line = line.replace('null','None')
        info = eval(line)
        yelp_data.append([info['name'], info['postal_code'], info['categories'], info['address']])
yelp_data
```

## OMNI

### Match by PlaceId:
 - filter yelp data by zipcodes in df
 - add google placeID to those zipcodes and match with df

#### Get mutual zipcodes

```python
omni_df = pd.read_csv('data/omni_place_details.tsv', sep='\t', index_col=0)

omni_df
```

```python
mutual_zips = list(set(omni_df.zipcode) & set(np.array(yelp_data)[:,1]))
yelp_data = np.array([x for x in yelp_data if x[1] in mutual_zips])
yelp_df = pd.DataFrame(yelp_data, columns=['name','zipcode','categories','address'])
yelp_df
```

#### get place_ids from yelp_data

```python
#key api key
with open('/Users/alexanderdaffara/Documents/api_info/ImpactRestaurantSearch/Impact_maps_api_key.txt', 'r') as f:
    key = f.readline()
pids = []
# Iterate over the restaurants and retrieve their Place IDs
for name, zipcode, address in zip(yelp_df.name, yelp_df.zipcode, yelp_df.address):
    # URL for Place Search API
    url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={name} {address} {zipcode}&inputtype=textquery&fields=place_id&key={key}'

    # Send a GET request to the API endpoint
    response = requests.get(url)

    # Check if the response was successful (HTTP status code 200)
    if response.status_code == 200:
        # Convert the response to JSON format
        data = response.json()

        # Extract the Place ID from the response
        if data['status'] == 'OK':
            place_id = data['candidates'][0]['place_id']
            pids.append(place_id)
        else:
            print(f'Request for {name} succeeded with status {data["status"]}')
            pids.append(np.nan)
    else:
        # Print an error message if the response was unsuccessful
        print(f'Request for {name} failed with error code {response.status_code}')
        pids.append(np.nan)
yelp_df['pid'] = pids
```

```python
merged_df = omni_df.merge(yelp_df, how='left', on='pid', suffixes=['',"_yelp"])
merged_df
```
